```python
class person:
  def __init__(self,name,age,school):
    self.name=name
    self.age=age
    self.school=school
  def __str__(self):
    return f"{self.name} is {self.age} and is schooling at {self.school}"
  def walk(self):
    print(f"{self.name} is walking")
  def aging(self):
    print(f"{self.name} has {self.age} yeears")
  def schooling(self):
    print(f"{self.name} is schooling at {self.school}")

rency=person("Rency",21,"KYU")
rency.walk()
rency.aging()
rency.schooling()

```
